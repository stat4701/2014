The problem here is that the data is clearly not normally distributed. This is due to the strategy that regulars take. About 75\% of the time (the complement to VP\$IP), they lose either 0, 0.25, or 0.5.

An approach is to bucket the hands together and leverage the CLT to make buckets that are normally distributed.

First we need to shuffle the hands to make sure there is no hidden bias:

```python
## STEP 1 - shuffle index and join as column
df = read_csv('5_ALL_HANDS.csv')
del df['AMT_WON']
shuffle_vals = df.index.values
np.random.shuffle(shuffle_vals)

shuffle = Series(shuffle_vals, index = range(len(df)))
shuffle.name = 'shuffle'

df = df.join(shuffle, how = 'left')
df.to_csv('6_SHUFFLE_STEP1.csv')

## STEP 2 - sort
df = read_csv('6_SHUFFLE_STEP1.csv', index_col = 'shuffle')
del df['Unnamed: 0']

df.sort_index(inplace = True)
df.to_csv('6_SHUFFLE_STEP2.csv')
```

We can use the 'shuffle' column above to bucket the hands. For example, if we had 500 hands and buckets of 50, then we'd have 10 buckets, and we'd use the modulo: 

${\tt bucket_i = ceiling\left(\frac{len(shuffled\_hands)}{bucket size}\right)\%shuffle_i = 10\%shuffle_i}$


```python
def all_buckets(df, bucket_size):
    total_buckets = np.ceil(len(df)/bucket_size)
    df['BUCKET'] = df.shuffle % total_buckets
    df_bucket = df.groupby('BUCKET').mean()
    del df_bucket['shuffle']
    return df_bucket
```

It's not the greatest fit but I think it's good enough that we can use it as our conjugate prior on the variance of player's winrates. scipy parameterizes its probability distributions differently than most texts or websites I've seen, so let's try to derive the scale and shape parameters.

In my experience, there is usually a bias term as one of the parameters.

```python
params = stats.invgamma.fit(all_players[all_players.HANDS>2500].EV_250_STD.values)
print params
```

```python
stats.invgamma.stats(params[0], params[1], params[2], 'mvsk')
```

In order, that's the mean, variance, skew, and kurtosis.

By wikipedia's parameterization:

$\alpha$ is shape, $\beta$ is scale

$mean = \frac{\beta}{\alpha-1} = 111.822$

$variance = \frac{\beta^2}{(\alpha-1)^2(\alpha-2)} = 662.40$

$skew = \frac{4\sqrt{\alpha-2}}{\alpha-3} = 0.416$

$kurtosis = \frac{30\alpha-66}{(\alpha-3)(\alpha-4)} = 0.327$


```python
alpha = params[0]
beta = params[2]
bias = params[1]

mean = beta/(alpha-1)+bias
variance = beta**2/((alpha-1)**2*(alpha-2))
skew = 4*sqrt(alpha-2)/(alpha-3)
kurtosis = (30*alpha-66)/((alpha-3)*(alpha-4))

print 'Mean: ' + str('%.3f' % mean)
print 'Variance: ' + str('%.3f' % variance)
print 'Skew: ' + str('%.3f' % skew)
print 'Kurtosis: ' + str('%.3f' % kurtosis)
```

In _Bayesian Data Analysis Third Edition_, Gelman et all parameterize the conjugate prior for the variance as an Inverse $\chi^2$ distribution (scipy does not support this distribution). That can be converted from an Inverse-Gamma as follows:

$\sigma^2 \sim Inv-\chi^2(\nu_0, \sigma_0^2)$, where $\nu = 2\alpha$

```python
nu_0 = 2*alpha
print 'nu_0: ' + str('%.3f' % nu)
```

$\mu|\sigma^2 \sim \mathcal{N}(\mu_0,\frac{\sigma^2_0}{\kappa_0})$

* $\kappa_0$ is the number of prior measurements in the data set
* $\mu_0$ is the sample mean of winrates in the data set
* $\sigma^2_0$ is the sample variance of the winrates in the data set

```python
df_conjugate = all_players[all_players.HANDS>2500]

# weighted mean and stdev
df_conjugate['TOT_EV'] = df_conjugate.HANDS*df_conjugate.EV_250/250 
df_conjugate['WEIGHT'] = df_conjugate.HANDS/df_conjugate.HANDS.sum()

mu_0 = df_conjugate.TOT_EV.sum() / df_conjugate.HANDS.sum() * 250

df_conjugate['WGT_SQR_ERR'] = (df_conjugate.EV_250 - mu_0)**2*df_conjugate.WEIGHT
sigma2_0 = df_conjugate.WGT_SQR_ERR.sum()

kappa_0 = len(df_conjugate)

print 'kappa_0: ' + str('%.3f' % kappa_0)
print 'mu_0: ' + str('%.3f' % mu_0)
print 'sigma2_0: ' + str('%.3f' % sigma2_0)
```

Per Gelman et al, the posterior parameters are as follows:

$\mu_n = \frac{\kappa_0}{\kappa_0+n}\mu_0 + \frac{n}{\kappa_0+n}\bar{y}$

$\kappa_n = \kappa_0 + n$

$\nu_n = \nu_0 + n$

$\nu_n\sigma^2_n = \nu_0\sigma_0^2 + (n-1)s^2 + \frac{\kappa_0n}{\kappa_0+n}(\bar{y}-\mu_0)^2$

$n, \bar{y}$, and $s^2$ are the number of buckets, the EV_250, and EV_250_STD, respectively, for each player.

```python
df_conjugate['n'] = df_conjugate.HANDS / 250
df_conjugate.ix[0]
```

```python
def kappa_p(kappa,n):
    return kappa+n

def mu_p(kappa,n,mu,ybar):
    return kappa*mu/kappa_p(kappa,n) + n*ybar/kappa_p(kappa,n)

def nu_p(nu,n):
    return nu+n

def sigma2_p(nu,sigma2,n,s,kappa,ybar,mu):
    part1 = nu*sigma2
    part2 = (n-1)*s**2
    part3 = (kappa*n)/(kappa+n)*(ybar - mu)**2
    return (part1 + part2 + part3) / nu_p(nu,n)

s = df_conjugate.EV_250_STD.ix[0]
ybar = df_conjugate.EV_250.ix[0]
n = df_conjugate.n.ix[0]

mu_n = mu_p(kappa_0,n,mu_0,ybar)
kappa_n = kappa_p(kappa_0,n)
nu_n = nu_p(nu_0,n)
sigma2_n = sigma2_p(nu_0,sigma2_0,n,s,kappa_0,ybar,mu_0)
```

```python
print mu_n
print sigma2_n/kappa_n
```

```python
df_conjugate['kappa_p'] = kappa_0 + df_conjugate.n
df_conjugate['nu_p'] = nu_0 + df_conjugate.n
df_conjugate['mu_p'] = kappa_0*mu_0/df_conjugate.kappa_p + df_conjugate.n*df_conjugate.EV_250/df_conjugate.kappa_p
df_conjugate['sigma2_p'] = ((nu_0*sigma2_0)+((df_conjugate.n-1)*df_conjugate.EV_250_STD**2)+
                            ((kappa_0*df_conjugate.n)/(kappa_0+df_conjugate.n)*
                             (df_conjugate.EV_250 - mu_0)**2))/(df_conjugate.nu_p)
df_conjugate['post_std'] = df_conjugate.sigma2_p/df_conjugate.kappa_p

df_conjugate = df_conjugate[['HANDS','EV_250','EV_250_STD','mu_p','post_std']]
df_conjugate.columns = 'HANDS','EV_250','EV_250_STD','EV_250_POST','EV_250_STD_POST'
```

We can also pick some players and see how their posterior distributions compare to their priors. In order to do this, we'll draw 100,000 random samples that each consist of:

 * Data: $\frac{34,500}{250}$ samples EV/250 and standard deviation
 * Posterior: 1 sample from posterior mean and standard deviation
 * Prior: 
   * Variance: 1 randomly drawn sample from the inverse gamma with our prior $\alpha$ and $\beta$,
   * Mean: prior mean (-0.56)

The posterior distribution is:

$p(\mu|y) = t_{\nu_n}\left(\mu|\mu_n,\frac{\sigma_n^2}{\kappa_n}\right)$

Again, scipy parameterizes their distributions in a different way. There are 3 parameters. We'll call them $t_0,t_1,t_2$:

$E[\mu|y] = t_1$

$Var[\mu|y] = \frac{t_0}{t_0-2}t_2^2$

Since we have $\mu_n$ and $\sigma^2_n$, we can derive the parameters:

$t_0 = \nu_n = \nu_0 + n$

$t_1 = E[\mu|y]$

$t_2 = \left(\frac{t_0-2}{t_0}\right)^{\frac{1}{2}}\sigma_n$

```python
def plot_player(player, xx, yy, h):

    n  = int(np.round(player.HANDS/250))
    
    data_actual = []
    data_posterior = []
    data_prior = []
    for i in range(10000):
        mean_actual = stats.norm.rvs(player.EV_250,player.EV_250_STD,size = n).mean()
        
        t_post = []
        t_post.append(player.HANDS/250+nu_0) #dof
        t_post.append(player.EV_250_POST) # bias
        t_post.append(sqrt((t_post[0]-2)*(player.EV_250_STD_POST**2)/(t_post[0])))
        
        mean_posterior = stats.t.rvs(t_post[0],t_post[1],t_post[2],size = 1)[0]

        sum_prior = 0
        for i in range(n):
            var_prior = stats.invgamma.rvs(alpha,beta,size = 1)[0]
            mean_prior = stats.norm.rvs(mu_0,var_prior / kappa_0,size = 1)[0]
            sum_prior += mean_prior

        data_actual.append(mean_actual)
        data_posterior.append(mean_posterior)
        data_prior.append(sum_prior/float(n))

    fig = plt.figure(figsize = (20,16))

    colors = brewer2mpl.get_map('Set2', 'qualitative', 4).mpl_colors

    ax = plt.subplot(311)
    ax.set_title('Hands Played: ' + str('%.0f' % player.HANDS) + ' | ' 
                 + 'EV/250: ' + str('%.2f' % player.EV_250) + ' | '
                 + 'STD/250: ' + str('%.2f' % player.EV_250_STD),
                 fontsize = 16, fontweight = 'bold', color = '#262626')

    binning = np.round((np.max(data_prior)-np.min(data_prior))/h)
    hist(data_prior, bins = binning, normed = False, 
         histtype = 'bar', cumulative=False, color = colors[0], alpha = .75, label = 'Prior')
    vlines(np.mean(data_prior), 0, yy[1], color = 'k', linestyle = '--')

    plt.xlabel('EV Won Per 250 Hands')
    plt.ylabel('Frequency')
    ax.legend(loc = 'best', fontsize = 20)

    for label in ax.get_xticklines():
        label.set_visible(False)
    for label in ax.get_yticklines():
        label.set_visible(False)

    xlim(xx)
    ylim(yy)

    ax = plt.subplot(312)
    binning = np.round((np.max(data_actual)-np.min(data_actual))/h)
    hist(data_actual, bins = binning, normed = False, 
         histtype = 'bar', cumulative=False, color = colors[1], alpha = .75, label = 'Data')
    vlines(np.mean(data_actual), 0, yy[1], color = 'k', linestyle = '--')

    plt.xlabel('EV Won Per 250 Hands')
    plt.ylabel('Frequency')
    ax.legend(loc = 'best', fontsize = 20)

    for label in ax.get_xticklines():
        label.set_visible(False)
    for label in ax.get_yticklines():
        label.set_visible(False)
    xlim(xx)
    ylim(yy)

    ax = plt.subplot(313)
    binning = np.round((np.max(data_posterior)-np.min(data_posterior))/h)
    hist(data_posterior, bins = binning, normed = False, 
         histtype = 'bar', cumulative=False, color = colors[2], alpha = .75, label = 'Posterior')
    vlines(np.mean(data_posterior), 0, yy[1], color = 'k', linestyle = '--')

    plt.xlabel('EV Won Per 250 Hands')
    plt.ylabel('Frequency')
    ax.legend(loc = 'best', fontsize = 20)

    for label in ax.get_xticklines():
        label.set_visible(False)
    for label in ax.get_yticklines():
        label.set_visible(False)

    xlim(xx)
    ylim(yy)


    plt.show()

plot_player(df_conjugate.ix[3], [-25,80], [0,3000], 1.5) #xlim, ylim, bin width
```

```python
def walk(player, length, num_walks):
    steps = int(np.ceil(length / 250))

    # make posterior parameters
    t_post = []
    t_post.append(player.HANDS/250+nu_0) #dof
    t_post.append(player.EV_250_POST) # bias
    t_post.append( sqrt ((t_post[0]-2) / t_post[0] * (player.EV_250_STD_POST*(kappa_0 + player.HANDS/250))))

    total_walk = {}
    for n in range(num_walks):
        this_walk = {}
        total_walk[n] = this_walk
        prev = 0
        hands = 250
        for i in range(steps):
            step = stats.t.rvs(t_post[0],t_post[1],t_post[2],size = 1)[0]
            location = step + prev
            this_walk[hands] = location
            prev = location
            hands += 250

    random_walk = DataFrame.from_dict(total_walk)
    return random_walk

player = df_conjugate.ix[3]
walk(player, 25000, 5)
print random_walk.head()
print random_walk.tail()
```
